# Introducción a Python

Python es un lenguaje de alto nivel de programación interpretado cuya filosofía hace hincapié en la legibilidad de su código. Se trata de un lenguaje de programación multiparadigma, ya que soporta parcialmente la orientación a objetos, programación imperativa y, en menor medida, programación funcional. (Fuente: Wikipedia)

Si bien tenemos esto claro de que Python es el más usado, y es bastante sencillo de aprender. ¿Qué características tiene Python que lo hacen ser el más usado?

- **Es un lenguaje de alto nivel**: Un lenguaje de alto nivel, es aquel que se asemeja más al lenguaje humano, por lo que su sintaxis es más fácil de memorizar. Python, es prácticamente como si a nuestra máquina le hablaramos en inglés.

- **Es un lenguaje orientado a objetos**: Son llamados "lenguajes orientados a objetos" a los lenguajes de programación que implementa los conceptos de la Programación Orientada a Objetos (POO).

- **Es un lenguaje Interpretado**: Python es un lenguaje que se ejecuta en tiempo real, línea por línea a través de un programa llamado intérprete. Esto permite identificar errores más rápido, ya que si hay alguno, el programa inmediatamente se detiene.

- **Portabilidad y multiplataforma**: Gracias a la característica anterior, Python es un lenguaje que se ejecuta en múltiples sistemas operativos y no hay necesidad de modificar el código.

- **Es de propósito general**: Gracias a paquetes desarrollados por la comunidad y con ayuda de las llamadas y creación de APIs, Python puede usarse en una gran cantidad de campos, por ejemplo, más adelante usaremos una librería llamada Reflex, que permite usar Python en desarrollo Web.

## Instalación de Python y Visual Studio Code

Para empezar a programar en Python requerimos de 2 elementos: El intérprete de Python que nos permitirá ejecutar nuestro código y un editor de código o IDE para poder escribir nuestro código.

Para instalar Python, debemos ir a la [Página Oficial de Python](https://www.python.org/) y descargar Python para nuestro sistema operativo, en mi caso Windows.

Paso 1: Ir a python.org e ir al apartado de descargas


Paso 2: Descargar Python para nuestro SO


Paso 3: Ejecutar el instalador e instalar Python



¡Bien! Ya tenemos Python en nuestra máquina, pero ahora, necesitamos una forma de escribir el código, y para eso usamos algo llamado editores de texto o IDEs. ¡Ojo! No debemos confundir uno con el otro. Un editor de texto es un programa liviano que nos permite modificar el texto o código de un archivo y nos ofrece algunas funciones adicionales, por ejemplo, Visual Studio Code. Pero un IDE es todo un entorno preparado con herramientas de desarrollo para un lenguaje en concreto, por ejemplo, para Python existe PyCharm.

En nuestro caso, usaremos el editor de texto Visual Studio Code y lo prepararemos con extensiones para convertirlo casi en un IDE para Python.

Para instalar Visual Studio Code debemos ir a la [Página Oficial de VSCode](https://code.visualstudio.com/) y descargarlo para nuestro sistema operativo. Cuando lo instalemos, debemos prepararlo con extensiones que nos permitirán desarrollar de una manera más cómoda en Python. Puede ver más información en la [Documentación De VSCode para Python](https://code.visualstudio.com/docs/python/python-quick-start)

Paso 1: Ir a code.visualstudio.com y descargar VSCode para nuestro SO


Paso 2: Ejecutar el instalador


Paso 3: Aceptar el acuerdo, crear un acceso directo e instalar VSCode (O solo denle "Siguiente" a todo)




Paso 4: Ejecutar VSCode e ir al apartado de extensiones


Paso 5: En el menú de extensiones, buscaremos las siguientes extensiones:

Python: Extensión de soporte para Python(instalará automáticamente Python Debugger y Pylance)



Prettier: Formateador de código



Python Indent: Indentador de código automático



Error Lens: Identifica en tiempo real errores en tu código



## Primer "¡Hola Mundo!" en Python

Una vez hayamos realizado las configuraciones anteriores, finalmente podemos empear a aprender Python de verdad. Lo primero que haremos será crear un nuevo archivo. Puedes hacerlo desde la pestaña "File" o directamente en el explorador.




A este archivo le darás un nombre cualquiera, pero el nombre debe terminar con la extensión de archivo `.py`, que es el formato de archivo de Python.



Una vez crees el archivo, se abrirá el editor de código, en el vamos a escribir lo siguiente:

```python
print("¡Hola Mundo!")
```

Le damos a "Save" en la pestaña "File" o pulsamos **Ctrl+S** en nuestro teclado para guardar el archivo, y luego ejecutaremos el código pulsando **F5** (iniciar depuración) o **Ctrl+F5** (ejecutar sin depurar). El resultado, debe ser que se abra una terminal mostrando el mensaje "¡Hola Mundo!"



¡Acabas de hacer tu primer Hola Mundo! ¡Bien!

# Jupyter Notebook

Jupyter Notebook es una aplicación web de documentos interactivos que contienen código y texto Markdown. Básicamente, combina código Python, multimedia y texto en un único archivo. Aunque Jupyter es ampliamente usado en la ciencia de datos, investigación y aprendizaje automático, puedes usarlo para cualquier proyecto que requiera documentación, como es el caso de este curso, que utiliza Jupyter porque permite que cada explicación esté acompañado de un ejemplo de código.

Si deseas utilizar Jupyter, puedes hacerlo desde VSCode con ayuda de una extensión:

Busca en el apartado de extensiones, la extensión llamada "Jupyter"



Ahora, crea un archivo con la extensión `.ipynb`



Cuando lo abras, tendrás un editor algo especial, ya que tiene otros botones y funciones y luce distinto al editor de texto convencional.
